So what is stereographic projection? 

Stereographic projection is a particular mapping (function) that projects a sphere onto a plane. The projection is defined on the entire sphere, except at one point: the projection point. Where it is defined, the mapping is smooth and bijective. 

In short:

1) We need to map all points of the sphere on a plane

2) Each point on the sphere can only correspond to 1 point on the plane

3) If the points are neighbours on the sphere, they are neighbours on the plane

So how do we acheive the above? The easiest way to do so is as follows:
![alt text](imagename.png "Projection")

1) Select a point  on the sphere (We are choosing the North pole)

2) Draw a line that passes through the North pole, the point we want to map and the plane

3) We will map the point to the point at which the line intersects with the plane

Let N = (0, 0, 1) be the "north pole", the points on the sphere be denoted by (x,y,z) and the points on the plane be denoted by (X,Y,0).

The plane z = 0 runs through the center of the sphere; the "equator" is the intersection of the sphere with this plane.

To determine the projection equation, consider the following:
For the line, when the z-component changes by (1-z) units, the x component changes by x units.
Thus, $$X = x/(1-z) $$
When the z-component changes by (1-z) units, the y component changes by y units.
Thus, $$Y = y/(1-z) $$

Hence, to plot the spherical projection,

```python
#import modules
import numpy as np
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# create the object 'figure'
fig = plt.figure()

#set the size so that it looks like a sphere
fig = plt.figure(figsize=plt.figaspect(1.0)*1.5) 
ax = fig.gca(projection='3d')

# draw the sphere (equation)
u, v = np.mgrid[0:2*np.pi:2000j, 0:np.pi:1000j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)

#function for stereographic projection
def stereo(x,y,z):
    c=1/(1.0000000000000001-z)
    return x+c*x,y+y*c
ax.plot_wireframe(x, y, z, color="b",linewidth=0.5)
plt.show()

figs = plt.figure()
figs = plt.figure(figsize=plt.figaspect(1.0)*1.5) 


a, b=stereo(x,y,z) #x+c*x,y+y*c , c=1/(1-z)
c, d=stereo(y,z,x) #y+c*y,z+z*c , c=1/(1-x)
e, f=stereo(z,x,y) #z+c*z,x+x*c , c=1/(1-y)
plt.scatter(a,b,c='blue',s=0.3)
plt.scatter(c,d,c='green',s=0.3)
plt.scatter(e,f,c='red',s=0.3)
plt.show()
```
